# 1. Create Lease on Chameleon (CHI@UC)

Reserve a `gpu_rtx_6000` bare-metal node at CHI@UC for serving evaluation.

**Lease name**: `Serving_Proj06`

```python
from chi import lease, context
import os

context.version = "1.0"
context.choose_project()
context.choose_site(default="CHI@UC")
```

```python
lease.add_node_reservation(
    lease_name="Serving_Proj06",
    node_type="gpu_rtx_6000",
    count=1
)
lease.set_start_date("now")
lease.set_end_date("+6h")

l = lease.create_lease("Serving_Proj06")
l.show()
```

```python
l = lease.get_lease("Serving_Proj06")
l.show()
```
