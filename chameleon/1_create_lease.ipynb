# 1. Create Lease on Chameleon (CHI@UC)

Reserve a `gpu_pi100` bare-metal node at CHI@TACC for serving evaluation.

**Lease name**: `Serving_Proj06`

```python
from chi import lease, context
import os

context.version = "1.0"
context.choose_project()
context.choose_site(default="CHI@TACC")
```

```python
from datetime import timedelta

l = lease.Lease(
    name="Serving_Proj06",
    duration=timedelta(hours=6)
)
l.add_node_reservation(node_type="gpu_p100", amount=1)
l.submit()
l.show()
```
